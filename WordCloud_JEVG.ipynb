```python
from wordcloud import WordCloud, ImageColorGenerator    # Hacer el wordcloud
from PIL import Image                                   # Cargar la imagen
import matplotlib.pyplot as plt                         # Graficar imagenes
import numpy as np                                      # Obtener la forma de la imagen

from stop_words import get_stop_words
Palabras_Irrelevanes = get_stop_words('spanish')
# Para quitar palabras manualmente:    Palabras_Irrelevanes.remove('Suelo')
Palabras_Irrelevanes

#Graficas el wordcloud
my_mask = np.array(Image.open('18.png'))
wc = WordCloud(background_color='white',
            collocations=False, # Desactivar palabras dobles
            #custom_stopwords=Palabras_Irrelevanes, # Quitar palabras irrelevantes
                mask=my_mask,
                    width=600,
                        height=300)

with open('Neumática-palabras.txt','r') as txt_file:
    texto = txt_file.read()

wc.generate(texto)

#Exstraer colores de imagen 
image_colors = ImageColorGenerator(my_mask)
wc.recolor(color_func=image_colors)

plt.figure(figsize=(20,10))      # Le das el tamaño
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')                   # Quitamos los bordes
wc.to_file('wordcloud18.png')     # Lo guardamos
plt.show()
```
